```python
import numpy as np
import numpy.linalg as npl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#https://datahub.io/core/sea-level-rise#resource-csiro_alt_gmsl_yr_2015
df=pd.read_csv('sea_level.csv')
df.head()
```

```python
plt.plot(np.sin(np.arange(100)*0.1))
plt.plot(np.cos(np.arange(100)*0.1))
plt.figure()
plt.plot(np.sin(np.arange(100)*0.2))
```

```python
plt.figure(figsize=(8,6))
plt.plot(df['Time'],df['GMSL'])
plt.title('Sea level change over time',fontsize=24)
plt.ylabel('Sea level change (mm)')
_=plt.xticks(ticks=df['Time'][::50])
plt.grid(True)
```

```python
y=np.array(df['GMSL'])
n=len(y)

X=np.ones((n,2))
X[:,0]=np.arange(n)
AB=npl.pinv(X).dot(y)
yhat=X.dot(AB)

plt.figure(figsize=(8,6))
plt.plot(df['Time'],df['GMSL'])
plt.plot(df['Time'],yhat)
plt.title('Sea level change over time',fontsize=24)
plt.ylabel('Sea level change (mm)')
_=plt.xticks(ticks=df['Time'][::50])
plt.grid(True)
```
